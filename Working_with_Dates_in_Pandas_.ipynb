```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
%matplotlib inline 
```

```python
# the first data set
person = pd.DataFrame({'Name':['Asif','Basit','John','Hary'],
                      'DOB': ['3/20/1960','3/19/1981','9/12/1999','7/12/1967'],
                       'EmpID': ['E453','E983','E675','E120']})
person
person.dtypes
```

```python
#Changes the data type of DOB coloumn to datetime
person['DOB']=pd.to_datetime(person['DOB'])
person
person.dtypes
```

```python
#Extracting month , date , year.....
person['Month']=person.DOB.dt.month
person['Day']=person.DOB.dt.day
person['Year']=person.DOB.dt.year
person['Day of Week']=person.DOB.dt.dayofweek
person['Month Name']=person.DOB.dt.month_name()
person['Day Name']=person.DOB.dt.day_name()
person
```

```python
# to get the current date
timestamp = pd.to_datetime('now')
print('Timestamp :{}'.format(timestamp))
```

```python
# how to get the age of the person
person['Age'] = pd.to_datetime('now').year - pd.to_datetime(person['DOB']).dt.year
person
```

```python
#Min Date frome the data frame..
pd.to_datetime(person['DOB']).min()
```

```python
#max date from the data frame
pd.to_datetime(person['DOB']).max()
```

```python
#covid data set to know more about describe func. 
from google.colab import drive
drive.mount('/content/drive')
covid = pd.read_csv('covid.csv')
covid.head(10)

covid.describe()
```

```python
#to know the travelling hours between starting date and ending date 
diff_hrs=(pd.to_datetime('2016-04-24 21:11:00')- pd.to_datetime('2016-04-23 20:11:00')).total_seconds()//3600
print('Date Difference in hours: {}' .format(diff_hrs))
```

```python
#to know the total trips done 
from datetime import date
uber1=uber[(pd.to_datetime(uber['START_DATE']) > pd.to_datetime(date(2016,1,14))) &
       (pd.to_datetime(uber['START_DATE']) < pd.to_datetime(date(2016,1,20)))]['trips'].sum()
print(uber1)
#for explanation
#import datetime
#d=datetime.date(2016,1,14)
#print(d)
```

```python
#min trips 
min_trip=uber['trips'].min()
min_trip
```

```python
#to know the min trips specific months 
wanted_output = uber.loc[uber["trips"] == 629 ,"Month_Name"]
wanted_output
```

```python
#max_trips
max_trip=uber['trips'].max()
max_trip
```

```python
#to know the mx trips in specific month
wanted_output = uber.loc[uber["trips"] == max_trip ,"Month_Name"]
wanted_output
```

```python
#Total number of trips  in each month using for loop
for i in uber['Month_Name'].unique():
    print('Total trips  in {0} : {1}' .format(i,uber[uber['Month_Name'] == i]['trips'].sum()))
```

```python
#making bar graph of trips

#making a proper order of month name
order=['January','February','March','April','May','June',
       'July','August','September','October','November','December']
uber['Month_Name']=pd.Categorical(uber['Month_Name'],order)

#grouping of the months by the "sum of trips"
uber.groupby('Month_Name').sum()[['trips']]

#finally making the bar grap of the grouped by data structure using matplotlib library
plt.figure(figsize=(14,6))
plt.bar(uber.groupby('Month_Name').sum().index,uber.groupby('Month_Name').sum()['trips'])
plt.show()
```

```python
# graph of covid usinf matplotlib library
plt.figure(figsize=(20,15))
plt.bar(covid.groupby('country').sum().index,covid.groupby('country').sum()['total_deaths'])
plt.show()
```

```python
#bar plot to get no of trips done in each month using pandas
uber.pivot_table(values=['trips'],index=['Month_Name'],aggfunc=np.sum).plot.bar()
```

```python
#Exporting a another data regarding female birth rate...
from google.colab import drive
drive.mount('/content/drive')
female = pd.read_csv('Female.txt')
female.head(10)
```

```python
#Change datatype of Data column to Datetime
female['Date']=pd.to_datetime(female['Date'])
female.dtypes
```

```python
# extracting day , month and year...
female['Month']=female.Date.dt.month
female['Day']=female.Date.dt.day
female['Month Name']=female.Date.dt.month_name()
female
```

```python
#we will convert "Month Name" column into Categorical variable
order=['January','February','March','April','May','June',
       'July','August','September','October','November','December']
female['Month Name']=pd.Categorical(female['Month Name'],order)

#using group by after ordering...
female.groupby('Month Name').sum()[['Births']]

#bar plot to get monthly female births using matplotlib 
plt.figure(figsize=(14,6))
plt.bar(female.groupby('Month Name').sum().index,female.groupby('Month Name').sum()['Births'])
plt.show()
```
